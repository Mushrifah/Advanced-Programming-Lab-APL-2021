## Count the number of lines in a file

```python
def count_lines(input_file):
    lines=0
    with open(input_file, 'r') as f:
        for line in f:
            line=line.split()
            lines= lines+ 1  
    print('Number of lines in the file is ',lines)
```

## Count the number of letters in a file

```python
def count_chars(input_file):
    lines=0
    with open(input_file, 'r') as f:
        f= f.read().replace(" ","")
        print('Number of characters in the file is ',len(f))
```
