<b>Aim Write a menu-driven text application to maintain bank accounts of customers
using Lists, Dictionary, Strings and Sets. This application handles error and exception
using try block. The application allows following operations as python functions: i) adding
and delete accounts of customers, ii) deposit and withdraw money to/from accounts, iii)
list last 3 transactions on a particular accounts and iv) exit from the application.</b>

## Add a customer

```python
cust=[]
def add_element():
    a = input("Enter the element to be added")
    cust.append(a)
    print("Customers in the bank are:",cust)
```

## Delete a customer

```python
def delete_element():
    b= input("Enter the element to be deleted: ")
    try:
        cust.remove(b)
        print("Updated customers in the bank are:",cust)
    except ValueError:
        print("Element not in the list")
```

## Create a customer dictionary with all transactions  

```python
d={}
def create_dict():
    key = str(input("Enter the name of customer: "))
    value = float(input("Enter the amount details of the customer: "))
    d.setdefault(key,[]).append(value)
    print(d)
    
```

## Deposit the money

```python
def deposit(d):
    try:
        k=input("Enter the name of customer")
        d[k]
    except:
        print("No such customer present in the bank")
    else:
        amt= float(input("Enter the amount to be deposited:- "))
        a=d[k][-1]
        f_amt= a + amt
        d.setdefault(k,[]).append(f_amt)
        print("Customer name: {name}\nCurrent amount: {amount}".format(name=k, amount=d[k][-1]))
        print(d)

    
```

## Withdraw the money

```python
def withdraw(d):
    try:
        k1=input("Enter the name of customer")
        d[k1]
    except:
        print("No such customer present in the bank")
    else:
        amt1= float(input("Enter the amount for withdrawal:- "))
        if (d[k1][-1] >= amt1 ):
            b= d[k1][-1]
            w_amt= b - amt1
            d.setdefault(k1,[]).append(w_amt)
        else:
            print("Insufficient balance")
        print("Customer name: {name}\nCurrent amount: {amount}".format(name=k1, amount=d[k1][-1]))
        print(d)
```

## Last 3 Transactions

```python
def transaction(d):
    try:
        name= input("Enter the customer name for last 3 transactions details:- ")
        list1= d[name]
        print("Last 3 transactions are:- ", list1[-3: ])
    except:
        print("No such customer name")
    
```
